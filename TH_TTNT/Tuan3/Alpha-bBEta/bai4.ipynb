```python
import math

player = 'X'
opponent = 'O'
N = 10
K = 5
MAX_DEPTH = 3

def is_moves_left(board):
    for i in range(N):
        for j in range(N):
            if board[i][j] == '_':
                return True
    return False

def evaluate(board):
    for i in range(N):
        for j in range(N - K + 1):
            if all(board[i][j+c] == player for c in range(K)):
                return 10
            if all(board[i][j+c] == opponent for c in range(K)):
                return -10

    for j in range(N):
        for i in range(N - K + 1):
            if all(board[i+c][j] == player for c in range(K)):
                return 10
            if all(board[i+c][j] == opponent for c in range(K)):
                return -10

    for i in range(N - K + 1):
        for j in range(N - K + 1):
            if all(board[i+c][j+c] == player for c in range(K)):
                return 10
            if all(board[i+c][j+c] == opponent for c in range(K)):
                return -10

    for i in range(K - 1, N):
        for j in range(N - K + 1):
            if all(board[i-c][j+c] == player for c in range(K)):
                return 10
            if all(board[i-c][j+c] == opponent for c in range(K)):
                return -10

    return 0

def alpha_beta(board, depth, is_max, alpha, beta):
    score = evaluate(board)

    if score == 10:
        return score - depth

    if score == -10:
        return score + depth

    if not is_moves_left(board) or depth >= MAX_DEPTH:
        return 0

    if is_max:
        best = -math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == '_':
                    board[i][j] = player
                    best = max(best, alpha_beta(board, depth + 1, not is_max, alpha, beta))
                    board[i][j] = '_'
                    alpha = max(alpha, best)
                    if beta <= alpha:
                        break
            if beta <= alpha:
                break
        return best
    else:
        best = math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == '_':
                    board[i][j] = opponent
                    best = min(best, alpha_beta(board, depth + 1, not is_max, alpha, beta))
                    board[i][j] = '_'
                    beta = min(beta, best)
                    if beta <= alpha:
                        break
            if beta <= alpha:
                break
        return best

def find_best_move(board):
    best_val = -math.inf
    best_move = (-1, -1)
    alpha = -math.inf
    beta = math.inf

    for i in range(N):
        for j in range(N):
            if board[i][j] == '_':
                board[i][j] = player
                move_val = alpha_beta(board, 0, False, alpha, beta)
                board[i][j] = '_'

                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val

    return best_move

board = [['_' for _ in range(N)] for _ in range(N)]

board[0][0] = 'X'
board[1][1] = 'O'
board[0][1] = 'X'

best_move = find_best_move(board)

print(f"Bàn cờ hiện tại ({N}x{N}):")
for row in board:
    print(row)

print("\nNước đi tối ưu nhất cho 'X':")
print("Hàng:", best_move[0])
print("Cột:", best_move[1])
```
