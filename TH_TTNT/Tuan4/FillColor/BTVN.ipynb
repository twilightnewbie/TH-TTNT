```python
import random

CLASSES = ['10A', '10B', '11A']
SUBJECTS = ['Toán', 'Tin']
TEACHERS = {
    'Toán': ['GV_Toán_1', 'GV_Toán_2'],
    'Tin': ['GV_Tin_1', 'GV_Tin_2']
}
SLOTS = list(range(20))

REQUIREMENTS = [
    ('10A', 'Toán'), ('10A', 'Toán'), ('10A', 'Tin'),
    ('10B', 'Toán'), ('10B', 'Toán'), ('10B', 'Tin'),
    ('11A', 'Toán'), ('11A', 'Tin'), ('11A', 'Tin')
]

POPULATION_SIZE = 100
GENERATIONS = 1000
MUTATION_RATE = 0.1

def create_individual():
    individual = []
    for req in REQUIREMENTS:
        subject = req[1]
        teacher = random.choice(TEACHERS[subject])
        slot = random.choice(SLOTS)
        individual.append((teacher, slot))
    return individual

def calculate_fitness(individual):
    penalty = 0
    class_slot_tracker = {}
    teacher_slot_tracker = {}

    for i in range(len(individual)):
        cls = REQUIREMENTS[i][0]
        teacher, slot = individual[i]

        class_slot = (cls, slot)
        if class_slot in class_slot_tracker:
            penalty += 1
        else:
            class_slot_tracker[class_slot] = 1

        teacher_slot = (teacher, slot)
        if teacher_slot in teacher_slot_tracker:
            penalty += 1
        else:
            teacher_slot_tracker[teacher_slot] = 1

    return 1.0 / (1.0 + penalty)

def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(REQUIREMENTS) - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

def mutate(individual):
    for i in range(len(individual)):
        if random.random() < MUTATION_RATE:
            subject = REQUIREMENTS[i][1]
            new_teacher = random.choice(TEACHERS[subject])
            new_slot = random.choice(SLOTS)
            individual[i] = (new_teacher, new_slot)
    return individual

def select_parents(population, fitnesses):
    total_fitness = sum(fitnesses)
    pick = random.uniform(0, total_fitness)
    current = 0
    for i in range(len(population)):
        current += fitnesses[i]
        if current > pick:
            return population[i]
    return population[-1]

def main():
    population = [create_individual() for _ in range(POPULATION_SIZE)]

    for generation in range(GENERATIONS):
        fitnesses = [calculate_fitness(ind) for ind in population]

        best_fitness = max(fitnesses)
        if best_fitness == 1.0:
            print(f"Đã tìm thấy lịch tối ưu tại thế hệ thứ {generation}!\n")
            break

        new_population = []
        while len(new_population) < POPULATION_SIZE:
            parent1 = select_parents(population, fitnesses)
            parent2 = select_parents(population, fitnesses)

            child1, child2 = crossover(parent1, parent2)

            child1 = mutate(child1)
            child2 = mutate(child2)

            new_population.extend([child1, child2])

        population = new_population[:POPULATION_SIZE]

    final_fitnesses = [calculate_fitness(ind) for ind in population]
    best_index = final_fitnesses.index(max(final_fitnesses))
    best_schedule = population[best_index]

    print("KẾT QUẢ SẮP XẾP LỊCH DẠY TỐI ƯU NHẤT:")
    print("-" * 60)

    for i in range(len(REQUIREMENTS)):
        cls, subject = REQUIREMENTS[i]
        teacher, slot = best_schedule[i]

        day = (slot // 4) + 2
        period = (slot % 4) + 1

        print(f"Lớp {cls:3} | Môn: {subject:4} | Giáo viên: {teacher:9} | Thứ {day}, Tiết {period}")

if __name__ == "__main__":
    main()
```
