```python
import sys
import os

def doc_ma_tran_chiphi(file_path):
    matrix = []
    if not os.path.exists(file_path):
        print(f"Lỗi: Không tìm thấy file {file_path}")
        return None
    with open(file_path, 'r') as f:
        for line in f:
            row_data = line.strip().replace(',', ' ').split()
            if row_data:
                matrix.append([int(x) for x in row_data])
    return matrix

def tsp_branch_and_bound(matrix):
    n = len(matrix)
    visited = [False] * n
    visited[0] = True

    best_path = []
    min_cost = sys.maxsize

    def backtrack(curr_city, count, current_cost, path):
        nonlocal min_cost, best_path

        if count == n and matrix[curr_city][0] > 0:
            total_cost = current_cost + matrix[curr_city][0]
            if total_cost < min_cost:
                min_cost = total_cost
                best_path = path + [0]
            return

        for next_city in range(n):
            if not visited[next_city] and matrix[curr_city][next_city] > 0:
                if current_cost + matrix[curr_city][next_city] < min_cost:
                    visited[next_city] = True

                    backtrack(next_city,
                              count + 1,
                              current_cost + matrix[curr_city][next_city],
                              path + [next_city])

                    visited[next_city] = False

    backtrack(0, 1, 0, [0])

    return best_path, min_cost

def main():
    ten_file = 'chiphi.txt'

    matrix = doc_ma_tran_chiphi(ten_file)
    if matrix is None:
        return

    so_thanh_pho = len(matrix)
    ten_thanh_pho = [chr(65 + i) for i in range(so_thanh_pho)]

    print(f"Bắt đầu tìm chu trình cho {so_thanh_pho} thành phố...\n")

    best_path, min_cost = tsp_branch_and_bound(matrix)

    if min_cost != sys.maxsize:
        hành_trình_chữ = [ten_thanh_pho[i] for i in best_path]
        chuoi_hanh_trinh = " -> ".join(hành_trình_chữ)

        print("KẾT QUẢ TỐI ƯU:")
        print(f"Hành trình đi: {chuoi_hanh_trinh}")
        print(f"Tổng chi phí tối thiểu: {min_cost}")
    else:
        print("Không tìm thấy chu trình nào đi qua tất cả các thành phố.")

if __name__ == "__main__":
    main()
```
