```python
import os

def doc_ma_tran_tu_file(file_path):
    """Hàm đọc ma trận kề từ file txt."""
    G = []
    if not os.path.exists(file_path):
        print(f"Lỗi: Không tìm thấy file {file_path}")
        return None

    with open(file_path, 'r') as f:
        for line in f:
            # Xóa khoảng trắng 2 đầu, thay dấu phẩy bằng khoảng trắng để xử lý linh hoạt
            # Phù hợp với cả định dạng cách nhau bằng phẩy hoặc khoảng trắng
            row_data = line.strip().replace(',', ' ').split()
            if row_data:
                # Chuyển các ký tự thành số nguyên
                row = [int(x) for x in row_data]
                G.append(row)
    return G

# --- CHƯƠNG TRÌNH CHÍNH ---

# 1. Đọc ma trận kề từ file txt bất kỳ
ten_file = 'matran.txt' # Đảm bảo bạn đã tạo file này cùng thư mục với code
G = doc_ma_tran_tu_file(ten_file)

if G is not None:
    so_dinh = len(G)

    # 2. Tự động sinh tên các đỉnh của đồ thị (A, B, C...) dựa trên số lượng đỉnh
    node = [chr(65 + i) for i in range(so_dinh)]

    t_ = {}
    for i in range(len(G)):
        t_[node[i]] = i

    # 3. Bậc của các đỉnh
    degree = []
    for i in range(len(G)):
        degree.append(sum(G[i]))

    # 4. Màu có thể sử dụng để tô cho các đỉnh (Mở rộng thêm nhiều màu)
    colorDict = {}
    tap_mau = ["Blue", "Red", "Yellow", "Green", "Purple", "Orange", "Pink", "Brown", "Cyan", "Magenta"]
    for i in range(len(G)):
        colorDict[node[i]] = tap_mau.copy()

    # 5. Sắp xếp các đỉnh theo thứ tự bậc (Sử dụng Selection Sort như bài mẫu)
    sortedNode = []
    indeks = []
    for i in range(len(degree)):
        _max = -1
        idx = -1
        for j in range(len(degree)):
            if j not in indeks:
                if degree[j] > _max:
                    _max = degree[j]
                    idx = j
        if idx != -1:
            indeks.append(idx)
            sortedNode.append(node[idx])

    # 6. Phần xử lý tô màu (sử dụng màu ít nhất có thể)
    theSolution = {}
    for n in sortedNode:
        setTheColor = colorDict[n]
        if not setTheColor:
            theSolution[n] = "Không đủ màu"
            continue

        theSolution[n] = setTheColor[0]
        adjacentNode = G[t_[n]]
        for j in range(len(adjacentNode)):
            if adjacentNode[j] == 1 and (setTheColor[0] in colorDict[node[j]]):
                colorDict[node[j]].remove(setTheColor[0])

    # 7. In kết quả từng đỉnh và màu đã tô tương ứng
    print("KẾT QUẢ TÔ MÀU ĐỒ THỊ:")
    for t, w in sorted(theSolution.items()):
        print(f"Đỉnh {t} = {w}")
```

```python
import os

# --- 1. CHƯƠNG TRÌNH CON: ĐỌC DỮ LIỆU ---
def doc_ma_tran_tu_file(file_path):
    """Đọc ma trận kề từ file txt."""
    G = []
    if not os.path.exists(file_path):
        print(f"Lỗi: Không tìm thấy file {file_path}")
        return None
    with open(file_path, 'r') as f:
        for line in f:
            row_data = line.strip().replace(',', ' ').split()
            if row_data:
                G.append([int(x) for x in row_data])
    return G

# --- 2. CHƯƠNG TRÌNH CON: TÍNH BẬC CÁC ĐỈNH ---
def tinh_bac_dinh(G):
    """Tính tổng bậc của từng đỉnh trong đồ thị."""
    degree = []
    for i in range(len(G)):
        degree.append(sum(G[i]))
    return degree

# --- 3. CHƯƠNG TRÌNH CON: SẮP XẾP ĐỈNH ---
def sap_xep_dinh_theo_bac(node, degree):
    """Sắp xếp các đỉnh theo thứ tự bậc giảm dần (sử dụng Selection Sort)."""
    sortedNode = []
    indeks = []
    for i in range(len(degree)):
        _max = -1
        idx = -1
        for j in range(len(degree)):
            if j not in indeks:
                if degree[j] > _max:
                    _max = degree[j]
                    idx = j
        if idx != -1:
            indeks.append(idx)
            sortedNode.append(node[idx])
    return sortedNode

# --- 4. CHƯƠNG TRÌNH CON: THUẬT TOÁN TÔ MÀU ---
def to_mau_do_thi(G, node, sortedNode):
    """Thực hiện tô màu đồ thị dựa trên mảng đỉnh đã sắp xếp."""
    # Mapping tên đỉnh với index
    t_ = {node[i]: i for i in range(len(G))}

    # Khởi tạo tập màu cho từng đỉnh
    tap_mau = ["Blue", "Red", "Yellow", "Green", "Purple", "Orange", "Pink", "Brown"]
    colorDict = {n: tap_mau.copy() for n in node}

    theSolution = {}
    for n in sortedNode:
        setTheColor = colorDict[n]
        if not setTheColor:
            theSolution[n] = "Không đủ màu"
            continue

        theSolution[n] = setTheColor[0]
        adjacentNode = G[t_[n]]

        # Cập nhật lại danh sách màu khả dụng của các đỉnh kề
        for j in range(len(adjacentNode)):
            if adjacentNode[j] == 1 and (setTheColor[0] in colorDict[node[j]]):
                colorDict[node[j]].remove(setTheColor[0])

    return theSolution

# --- CHƯƠNG TRÌNH CHÍNH (MAIN) ---
def main():
    ten_file = 'matran.txt'

    # B1: Đọc đồ thị
    G = doc_ma_tran_tu_file(ten_file)
    if G is None:
        return

    # Khởi tạo danh sách đỉnh tự động (A, B, C, D,...)
    so_dinh = len(G)
    node = [chr(65 + i) for i in range(so_dinh)]

    # B2: Tính bậc
    degree = tinh_bac_dinh(G)

    # B3: Sắp xếp
    sortedNode = sap_xep_dinh_theo_bac(node, degree)

    # B4: Tô màu
    ket_qua = to_mau_do_thi(G, node, sortedNode)

    # In kết quả
    print("KẾT QUẢ TÔ MÀU ĐỒ THỊ:")
    for t, w in sorted(ket_qua.items()):
        print(f"Đỉnh {t} = {w}")

if __name__ == "__main__":
    main()
```
