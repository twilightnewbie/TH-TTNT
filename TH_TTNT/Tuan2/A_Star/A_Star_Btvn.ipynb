```python
import heapq

def get_mst_cost(graph, unvisited):
    if not unvisited:
        return 0
    unvisited_list = list(unvisited)
    n = len(unvisited_list)
    if n == 1:
        return 0

    in_mst = {unvisited_list[0]}
    mst_cost = 0

    while len(in_mst) < n:
        min_edge = float('inf')
        best_node = None
        for u in in_mst:
            for v in unvisited:
                if v not in in_mst and graph[u][v] < min_edge:
                    min_edge = graph[u][v]
                    best_node = v
        mst_cost += min_edge
        in_mst.add(best_node)

    return mst_cost

class TSPNode:
    def __init__(self, current, visited, path, g, h):
        self.current = current
        self.visited = frozenset(visited)
        self.path = path
        self.g = g
        self.h = h
        self.f = g + h

    def __lt__(self, other):
        return self.f < other.f

def solve_tsp_astar(graph, start=0):
    n = len(graph)
    initial_visited = {start}

    unvisited_initial = set(range(n)) - initial_visited
    h_initial = get_mst_cost(graph, unvisited_initial)

    if unvisited_initial:
        h_initial += min(graph[start][v] for v in unvisited_initial)
        h_initial += min(graph[v][start] for v in unvisited_initial)

    start_node = TSPNode(start, initial_visited, [start], 0, h_initial)

    open_list = []
    heapq.heappush(open_list, start_node)

    while open_list:
        current_node = heapq.heappop(open_list)

        if len(current_node.visited) == n:
            final_cost = current_node.g + graph[current_node.current][start]
            return current_node.path + [start], final_cost

        unvisited = set(range(n)) - current_node.visited

        for next_city in unvisited:
            next_visited = current_node.visited | {next_city}
            next_path = current_node.path + [next_city]
            next_g = current_node.g + graph[current_node.current][next_city]

            rem_unvisited = set(range(n)) - next_visited

            h_next = get_mst_cost(graph, rem_unvisited)
            if rem_unvisited:
                h_next += min(graph[next_city][v] for v in rem_unvisited)
                h_next += min(graph[v][start] for v in rem_unvisited)
            else:
                h_next = graph[next_city][start]

            next_node = TSPNode(next_city, next_visited, next_path, next_g, h_next)
            heapq.heappush(open_list, next_node)

    return None, float('inf')

if __name__ == "__main__":
    graph = [
        [0, 10, 15, 20],
        [10, 0, 35, 25],
        [15, 35, 0, 30],
        [20, 25, 30, 0]
    ]

    path, cost = solve_tsp_astar(graph, start=0)
    print("Đường đi tối ưu:", path)
    print("Tổng chi phí (g):", cost)
```
