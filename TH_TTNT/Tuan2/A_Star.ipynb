```python
class Graph:
    def __init__(self, adjac_lis, heuristic_dict):
        self.adjac_lis = adjac_lis
        self.H = heuristic_dict
    def get_neighbors(self, v):
        return self.adjac_lis.get(v, [])
    def h(self, n):
        return self.H.get(n, 0)
    def a_star_algorithm(self, start, stop):
        open_lst = set([start])
        closed_lst = set([])
        g = {}
        g[start] = 0
        parents = {}
        parents[start] = start

        while len(open_lst) > 0:
            n = None
            for v in open_lst:
                if n is None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v
            if n is None:
                print('Không tìm thấy đường đi!')
                return None
            if n == stop:
                reconst_path = []
                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]
                reconst_path.append(start)
                reconst_path.reverse()
                print(f'Đường đi từ {start} đến {stop}: {reconst_path}')
                return reconst_path
            for (m, weight) in self.get_neighbors(n):
                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n

                        if m in closed_lst:
                            closed_lst.remove(m)
                            open_lst.add(m)
            open_lst.remove(n)
            closed_lst.add(n)
        print('Không tìm thấy đường đi!')
        return None
adjac_lis = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('C', 1), ('D', 5)],
    'C': [('D', 12)]
}
heuristic_dict = {
    'A': 1,
    'B': 1,
    'C': 1,
    'D': 0
}

graph = Graph(adjac_lis, heuristic_dict)
print("--- TEST CASE 1 ---")
graph.a_star_algorithm('A', 'C')
print("\n--- TEST CASE 2 ---")
graph.a_star_algorithm('A', 'D')
print("\n--- TEST CASE 3 ---")
graph.a_star_algorithm('B', 'D')
```
